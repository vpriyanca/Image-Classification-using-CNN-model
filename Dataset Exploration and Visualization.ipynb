```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense,Input,Dropout,GlobalAveragePooling2D,Flatten,Conv2D,BatchNormalization,Activation,MaxPooling2D
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.optimizers import Adam,SGD,RMSprop
```

```python
picture_size = 48
folder_path = r"/Users/vpriyanca/Downloads/ML Project_1_2/images_IC/images"
```

```python
classes = 'Class_1_Touch'

plt.figure(figsize= (12,12))
for i in range(1, 10, 1):
    plt.subplot(3,3,i)
    img = load_img(folder_path +"/train/"+ classes +"/"+
                  os.listdir(folder_path + "/train/" + classes)[i], target_size=(picture_size, picture_size))
    plt.imshow(img)   
plt.show()
```
